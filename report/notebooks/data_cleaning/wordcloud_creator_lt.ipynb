```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

lem = pd.read_csv('../../streamlit_app/lem_example.csv')
stem = pd.read_csv('../../streamlit_app/stem_example.csv')
```

```python
import sys
sys.path.append('./lem_stem_functions')

from text_functions_new_vocabs_ac import new_column_lemmatizer, new_column_stemmatizer, new_count_vectorize_data, new_tfidf_vectorize_data

df = pd.read_json("../../../data/raw/Appliances.json", lines = True)
df = df.dropna(subset='reviewText')

features = df['reviewText'].iloc[:50000]

lemmed_features = new_column_lemmatizer(features)
```

```python
# Combine the lemmatized text from the Series into a single string
combined_text = " ".join(lemmed_features.explode().astype(str))
# Generate the WordCloud
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(combined_text)
# Plot the WordCloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud Visualization")
# plt.savefig('../images/WordcloudLem2.png')
plt.show()
```
